```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,100)
print(x)
```

```python
y = np.sin(x)
print(y)
```

```python
z = np.cos(x)
print(z)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r')
plt.show()
```

```python
plt.plot(x,y,'0.8')
plt.show()
```

```python
plt.plot(x,y,'cyan')
plt.show()
```

```python
plt.plot(x,y,'b')
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,y,'+b')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ["English","French","Spanish","Latin","German"]
people = [100,10,150,40,70]
ax.bar(languages,people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]
ax.pie(people, labels=languages, autopct='%0.1f%%')
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='b')
plt.show()



```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z= 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
